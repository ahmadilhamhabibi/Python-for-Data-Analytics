### Import Pandas dan Import Dataset

```python
import pandas as pd

df = pd.read_csv("datasets/world_population.csv")

df.head()
```

### Filter kolom tertentu yang ingin ditampilkan

```python
df = df[["Rank", "CCA3", "Country/Territory", "Capital", "Continent", "2022 Population", 
         "Area (km²)", "Density (per km²)", "Growth Rate", "World Population Percentage"]]

df.head()
```

#### Mengubah nama kolom

```python
df = df.rename(columns={"Country/Territory": "Country" , 
                        "2022 Population": "Population"})

df.head()
```

### Filter data menggunakan operasi matematika

- sama dengan ( == )
- tidak sama dengan ( != )
- lebih dari ( > )
- lebih dari sama dengan ( >= )
- kurang dari ( < )
- kurang dari sama dengan ( <= )

```python
df[df.Country == "Indonesia"]
```

```python
df[df["Population"] == 11212191 ]
```

```python
df[df["Continent"] != "Asia" ].head(10)
```

```python
df[df["Population"] > 215313498 ].sort_values("Population", ascending=False)
```

```python
df[df["Population"] >= 215313498 ].sort_values("Population", ascending=False)
```

```python
df[df["Area (km²)"] < 30 ].sort_values("Area (km²)")
```

```python
df[df["Area (km²)"] <= 30 ].sort_values("Area (km²)")
```

## Import Seaborn and Load Datasets

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

## Scatter Plot

```python
sns.scatterplot(data=tips, x="tip", y="total_bill", hue="day", size="size", palette="YlGnBu")
```

## Histogram and Distribution Plots

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

```python
sns.displot(tips['tip'], kde=True, bins=15)
```

## Bar Plot

```python
sns.barplot(data=tips, x="sex", y="tip", palette="YlGnBu")
```

## Box Plot and Strip Plot

```python
sns.boxplot(data=tips, x="day", y="tip", hue="sex", palette="YlGnBu")
```

```python
sns.stripplot(data=tips, x="day", y="tip", hue="sex", palette="YlGnBu", dodge=True)
```

## Join Plot

```python
sns.jointplot(data=tips, x="tip", y="total_bill", kind="hex", palette="YlGnBu")
```

## Pair plot

```python
sns.pairplot(titanic.select_dtypes(['number']), hue="pclass", palette="YlGnBu")
```

## Heat Map

```python
titanic.corr()
```

```python
sns.heatmap(titanic.corr(), annot=True, cmap="YlGnBu")
```

## Cluster Map

```python
sns.clustermap(iris.drop("species", axis=1), cmap="YlGnBu")
```

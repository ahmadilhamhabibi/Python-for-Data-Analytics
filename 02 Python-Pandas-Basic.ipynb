## df.sort_values() 
#### Digunakan untuk mengurutkan data berdasarkan kolom tertentu. By default akan mengurutkan nilai dari terendah (ascending)

```python
# data awal sebelum diurutkan
df.head()
```

```python
# melihat 10 negara dengan rangking (RANK) tertinggi
df.sort_values("Rank").head(10)
```

#### Jika ingin mengurutkan dari nilai tertinggi kita gunakan parameter **ascending=False**

```python
# melihat 10 negara dengan rangking (RANK) terendah
df.sort_values("Rank", ascending=False).head(10)
```

#### Mengurutkan data berdasarkan banyak kolom

```python
df.sort_values(["Continent", "2022 Population"], ascending=[True, False]).head(10)
```

```python
df[["Rank", "Country/Territory", "Continent"]].sort_values(["Continent", "Country/Territory"], ascending=[True, True])
```

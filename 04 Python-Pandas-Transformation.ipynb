### Import Pandas dan Import Dataset

````python
import pandas as pd

df = pd.read_csv("datasets/world_population.csv")

df.info()
````

#### Menghapus beberapa kolom 

````python
df = df.drop(columns=["2020 Population", "2020 Population", "2015 Population", "2010 Population", "2000 Population",
                      "1990 Population", "1980 Population", "1970 Population"])

df.head()
````

#### Mengubah nama kolom

````python
df = df.rename(columns={"Country/Territory": "Country",
                        "2022 Population": "Population"})
df.head()
````

````python
df.info()
````

## Mengubah tipe data pada kolom

Dengan perintah ```df.info()``` kita bisa mendapatkan informasi tentang tipe data pada masing-masing kolom. 
Jika dirasa tipe datanya kurang tepat, kita bisa mengubah tipe data sesuai keinginana kita
dengan memanfaatkan perintah ```.astype(str/int/float)```

````python
df1 = df.copy()
````

````python
# Mengubah kolom Rank menjadi tipe String
df1.Rank = df1.Rank.astype(str)


# Mengubah kolom Population menjadi Float
df1["Population"] = df1["Population"].astype(float)


# Mengubah kolom World Population Percentage menjadi Integer
df1["World Population Percentage"] = df1["World Population Percentage"].astype(int)


df1.info()

df1.head()
````

## Mengubah teks 

- **.str.lower()** untuk mengubah teks menjadi huruf kecil (abcde)
- **.str.upper()** untuk mengubah teks menjadi huruf besar (ABCDE)
- **.str.title()** untuk mengubah teks menjadi *title case* (Abcde)

````python
df.CCA3 = df.CCA3.str.lower()

df.head()
````

````python
df["CCA3"] = df["CCA3"].str.title()

df.head()
````

````python
df["CCA3"] = df["CCA3"].str.upper()

df.head()
````

## Filter berdasarkan pola tertentu

- **.str.contains(...)** memfilter menggunakan pola tertentu
- **.str.contains(^..)** memfilter menggunakan pola tertentu yang diawali dengan karakter yang dicari
- **.str.contains(..$)** memfilter menggunakan pola tertentu

````python
# Menampilkan semua negara yang terdapat kata "tin"
df[df["Country"].str.contains("tin")]
````

````python
# Menampilkan semua negara yang dengan awalan "Ir"

df[df["Country"].str.contains("^Ir")]
````

````python
# Menampilkan semua negara yang dengan akhiran "sia"
df[df["Country"].str.contains("sia$")]
````

### untuk data numerik tetap bisa memanfaatkan fitur .str.countains dengan cara diubah menjadi tipe string terlebih dahulu

````python
df[df["World Population Percentage"].astype(str).str.contains("^2")]
````

## Menambah Kolom

Untuk menambah kolom bisa dilakukan dengan cara ```df[kolom_baru] = list ```

Kita juga bisa menambahkan kolom baru dengan menggabungankan kolom-kolom yang sudah ada.

````python
df["Code Country"] = df["CCA3"]+" - "+df["Country"]

df.head(10)
````

````python
# Kita bisa membuat kolom baru dengan operator matematika ( +, -, *, / )
df["Area/Population"] = df["Area (km²)"] / df["Population"]

df.head()
````

## Menghapus Kolom dan Menghapus Baris

Seperti yang telah kita lakukan diatas, kita bisa menghapus kolom dengan ```df.drop()```
Untuk menghapus baris tertentu, kta gunakan filter untuk menentukan data yang ingin dihapus

````python
df = df.drop(columns=["Code Country", "Area/Population"])

df.head()
````

````python
# Membuat data baru untuk menghapus baris
df2 = df.copy()

df2 = df2.sort_values("Rank")
df2.head()
````

````python
# menghapus data dengan populasi lebih dari 275501339
df2 = df2[~(df2["Population"] > 275501339)]
df2.head()
````

````python
# Menghapus kolom dengan nama negara yang terdapat kata "ia"
df2 = df2[~(df2["Country"].str.contains("ia"))] # menghapus Indonesia dan Nigeria

df2.head()
````
